Cocktail DB Example

```python
url = 'https://www.thecocktaildb.com/api/json/v1/1/filter.php?c=Cocktail'

r = requests.get(url)
r
```

```python
r.text
```

```python
#r.json
myjson = json.loads(r.text)
myjson
```

```python
myjson['drinks'][0]['strDrinkThumb']
```

** Records: the list of equivalently formatted dictionaries that you want to become ROWS in a dataframe  
** Record path: the sequence of JSON keys you need to delve into to gt to the records

In this cas the record path is just 'drinks'

```python
cocktail_df = pd.json_normalize(myjson, record_path=['drinks'])
cocktail_df
```

Wikipedia API

```python
url = 'https://en.wikipedia.org/w/api.php?action=query&list=search&srsearch=Virginia&format=json&srlimit=500'
r = requests.get(url)
r
```

```python
mysjon = r.json()
wiki_df = pd.json_normalize(mysjon, record_path=['query', 'search'])
wiki_df
```
