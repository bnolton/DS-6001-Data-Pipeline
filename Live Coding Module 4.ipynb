```python
import numpy as np
import pandas as pd
import requests
import json
import dotenv
import os
```

A "dot E N V" file, named .env, is a plain text file for storing passwords/keys/secrets

Because it begins with a . for a file name, it is understood by your computer to be a HIDDEN file. That's extra security.

When using GitHub, always us a .gitignore file. .env will always be one of the ignored files. DO NOT allow you .env file to be uploaded to GitHub.

TO load secret keys from the .env file

```python
dotenv.load_dotenv()
newskey = os.getenv('newskey')
```

```python
url = 'https://newsapi.org/v2/everything'
params = {'apiKey': newskey,
          'q': 'luka'}
r = requests.get(url, params=params) 
myson = r.json()
news_df = pd.json_normalize(myson, record_path=['articles'])
news_df
```

Congress API

```python
congresskey = os.getenv('congresskey')
```

User agent strings

User agent string is a string that informs the API about what kind of system you are using to access the API. It's considered good etiquite to include one, and it is sometimes required by the API

```python
url = 'https://httpbin.org/user-agent'
r = requests.get(url)
myjson = r.json()
useragent = myjson['user-agent']
```

```python
headers = {'User-Agent': useragent,
           'From': 'frv3fp@virginia.edu'}
```

Four things to look for in API documentation:
1. The root
2. The endpoint for the data source you need
3. The parameters
4. Headers, or other instructions for supplying your API key

```python
root = 'https://api.congress.gov/v3' #the begging of the all URLs attached to this API
endpoint = '/member' # the next par of that URL that leads to a particular resource on the API
params = {'format': 'json',
          'limit': '250',
          'offset': '0',
          'currentMember': 'true',
          'apikey': congresskey}
r = requests.get(root + endpoint, params=params, headers=headers)
r
```

```python
myjson = json.loads(r.text)
```

```python
congress_df = pd.json_normalize(myjson, record_path=['members'])
congress_df
```

FEC

```python
dotenv.load_dotenv()
feckey = os.getenv('feckey')
```

```python
root = 'https://api.open.fec.gov/'
endpoint = 'v1/candidates'
params = {'api_key': feckey,
          'q': 'mcguire'}
r = requests.get(root + endpoint, params=params)
r
```
